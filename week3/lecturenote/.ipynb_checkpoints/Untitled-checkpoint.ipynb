```python
from math import sqrt
def generate_all_primes_upto(n):
    sieve=[True]*(n+1)
    
    for i in range(2,round(sqrt(n))+1):
        d=i**2
        while d <= n:
            sieve[d] =False
            d+=i
    for i in range(2,n+1):
        if sieve[i]:
            print(i)
        
generate_all_primes_upto(1000000)
```

```python
def generate_all_prime(n):
    list_of_primes=list(range(2,n+1))
    i=0
    while list_of_primes[i] <=round(sqrt(n)):
        k=0
        factor =list_of_primes[i]*list_of_primes[i+k]
        while True:
            
            factor =list_of_primes[i]*list_of_primes[i+k]
            while factor <=n:
                list_of_primes.remove(factor)
                factor *= list_of_primes[i]
            k+=1
        i+=1

        
```

```python
def primes_break(denominator1):
    
    if denominator1 == 1:
        return True
    else:
        primes_num=[]
        for k in range(2,denominator1+1):
            count=0
            while denominator1%k == 0:
                denominator1 /= k
                count+=1
            if count>0:
                for each2 in range(count):
                    primes_num.append(k)
        primes_num_len=len(primes_num)
        count_len=0
        for each1 in primes_num:
            if each1 % 2 ==0 or each1 % 5 ==0:
                count_len+=1
        if count_len == primes_num_len:
            return True
        else:
            return False
primes_break(1)
```
